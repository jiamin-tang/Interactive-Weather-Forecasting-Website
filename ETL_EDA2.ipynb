```python
import requests
import datetime
import pandas as pd
```

```python
def load_forecast_data(location):
    '''
    dates: location
    return: df_daily_forecast
            df_hourly_forecast
    '''
    interval = 1
    url = 'http://api.worldweatheronline.com/premium/v1/weather.ashx'
    key = '8fefb61db8a241c4b7524155190612'
    fm = 'json'
    
    num_of_days = 7
    show_comments = 'no'
    show_local_time = 'yes'
    show_map = 'yes'
    
    
    paras = dict()
    paras['key'] = key
    paras['format'] = fm
    paras['q'] = location
    paras['tp'] = interval
    paras['num_of_days'] = num_of_days
    paras['show_comments'] = show_comments
    paras['showlocaltime'] = show_local_time

        
    r = requests.get(url, paras)
    r.raise_for_status()
    return r.json()
```

```python
res = load_forecast_data('providence')
```
